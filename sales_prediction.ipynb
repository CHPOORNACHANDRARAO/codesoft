```python
!pip install scikit-learn # install scikit-learn if it's not already installed

from sklearn.model_selection import train_test_split # import the train_test_split function

x=sales[['TV']]
y=sales[['Sales']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
ML=LinearRegression()
ML.fit(x_train,y_train)
```

```python
x_train_predicition=ML.predict(x_train)
print(x_train_predicition)
```

```python
print(y_train)
```

```python
print(ML.coef_)
print(ML.intercept_)

```

```python
x_test_predicition=ML.predict(x_test)
print(x_test_predicition)
```
